```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")

human1 = Human("田中", 25)
human2 = Human("佐藤", 18)
human3 = Human("鈴木", 20)

humans = [human1, human2, human3]

for human in humans:
    human.check_adult()
```
